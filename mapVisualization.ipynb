```python
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

#varibles

```python
figureSize = (5, 5)
fontSize = figureSize[0]

outlineColor = 'black'
backgroundColor = 'lightcyan'
colorMap = 'jet'
custom_cmap = mcolors.ListedColormap(['#FF5733', '#33FF57', '#3357FF', '#FFFF33', '#FF33FF', '#33FFFF'])
custom_cmap_gradient = mcolors.LinearSegmentedColormap.from_list("custom_gradient", ["red", "yellow", "green", "blue"])

lineWidth = 1
```

#data

```python
bangladesh = gpd.read_file("https://raw.githubusercontent.com/DryBoss/map-visualization/main/datas/whole_bangladesh_outline.json")
bangladesh_divisions = gpd.read_file("https://raw.githubusercontent.com/DryBoss/map-visualization/main/datas/divisions_bibhags_outline.json")
bangladesh_districts = gpd.read_file("https://raw.githubusercontent.com/DryBoss/map-visualization/main/datas/districts_zillas_outline.json")
bangladesh_subDistricts = gpd.read_file("https://raw.githubusercontent.com/DryBoss/map-visualization/main/datas/upozila_outline.json")
bangladesh_unions = gpd.read_file("https://raw.githubusercontent.com/DryBoss/map-visualization/main/datas/unions_thanas_outline.json")
```

# Bangladesh

```python
fig,ax = plt.subplots(figsize=figureSize)
ax = bangladesh.plot(ax=ax, edgecolor=outlineColor, color=backgroundColor, linewidth=lineWidth)
ax.set_title('Bangladesh')
```

##divisions

```python
fig,ax = plt.subplots(figsize=figureSize)
ax = bangladesh_divisions.plot(ax=ax , edgecolor=outlineColor, color=backgroundColor, linewidth=lineWidth)
ax.set_title('Divisions of Bangladesh')
```

```python
fig,ax = plt.subplots(figsize=figureSize)
ax = bangladesh_divisions.plot(ax=ax, edgecolor=outlineColor, cmap=colorMap, column='ADM1_EN', linewidth=lineWidth)
ax.set_title('Divisions in Bangladesh with different color')
```

```python
fig,ax = plt.subplots(figsize=figureSize)
ax = bangladesh_divisions.plot(ax=ax, edgecolor="black", cmap=colorMap, column='ADM1_EN', linewidth=lineWidth)
bangladesh_divisions.apply(lambda x: ax.annotate(text=x.ADM1_EN, xy=x.geometry.centroid.coords[0], ha='center', fontsize=fontSize), axis=1);
ax.set_title('Name of Divisions in Bangladesh with different color')
```

```python
fig, ax = plt.subplots(figsize=figureSize)
ax = bangladesh_divisions.plot(ax=ax, edgecolor="black", cmap=custom_cmap_gradient, column="ADM1_EN", linewidth=lineWidth)
ax.set_title('Divisions in Bangladesh with a Custom Gradient')
```
